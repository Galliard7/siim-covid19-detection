# Import libraries

```python
import numpy as np
import pandas as pd
```

# Collect OOF files

```python
train_image_df = pd.read_csv("../input/siim-covid19-detection/train_image_level.csv")
train_study_df = pd.read_csv("../input/siim-covid19-detection/train_study_level.csv")
train_study_df['StudyInstanceUID'] = train_study_df['id'].apply(lambda x: x.split('_')[0])

train_df = pd.merge(train_image_df, train_study_df, on="StudyInstanceUID", how="left")

train_df.rename(columns={'id_x':'image_id',
                         'id_y':'study_id'}, inplace=True)

train_df['image_file_id'] = train_df['image_id'].apply(lambda x: x.split('_')[0])

train_df.head(3)
```

```python
efnb7_df = pd.read_csv("../input/siimcovid19trainedmodelsstudyclassification/efnb7_oof.csv")
swin_df = pd.read_csv("../input/siimcovid19trainedmodelsstudyclassification/swin_valid.csv")
eff3_df = pd.read_csv("../input/siimcovid19trainedmodelsstudyclassification/tf_eff3_valid.csv")
effv2_df = pd.read_csv("../input/siimcovid19trainedmodelsstudyclassification/effv2l_oof_no_tta.csv")
effv2_tta1_df = pd.read_csv("../input/siimcovid19trainedmodelsstudyclassification/effv2l_oof/effv2l_oof/effv2l_oof_1_tta.csv")
effv2_tta2_df = pd.read_csv("../input/siimcovid19trainedmodelsstudyclassification/effv2l_oof/effv2l_oof/effv2l_oof_2_tta.csv")
effv2_tta4_df = pd.read_csv("../input/siimcovid19trainedmodelsstudyclassification/effv2l_oof/effv2l_oof/effv2l_oof_4_tta.csv")
effv2_tta6_df = pd.read_csv("../input/siimcovid19trainedmodelsstudyclassification/effv2l_oof/effv2l_oof/effv2l_oof_6_tta.csv")

print("efnb7")
print(efnb7_df.shape)
print()

print("swin")
print(swin_df.shape)
print()

print("eff3")
print(eff3_df.shape)
print()

print("effv2")
print(effv2_df.shape)
print()

print("effv2 tta1")
print(effv2_tta1_df.shape)
print()

print("effv2 tta2")
print(effv2_tta2_df.shape)
print()

print("effv2 tta4")
print(effv2_tta4_df.shape)
print()

print("effv2 tta6")
print(effv2_tta6_df.shape)
print()
```

```python
temp = train_df[['study_id','image_file_id']].copy()

efnb7_df = pd.merge(efnb7_df, temp, left_on="id", right_on="study_id")

efnb7_df.rename(columns={'image_file_id':'image_id',
                         'negative':'efnb7_neg',
                         'typical':'efnb7_typ',
                         'indeterminate':'efnb7_ind',
                         'atypical':'efnb7_atyp',}, inplace=True)

efnb7_df.drop(columns=['id'], inplace=True)

print(efnb7_df.shape)
efnb7_df.head(2)
```

```python
swin_df.rename(columns = {'Unnamed: 0':'image_id'}, inplace=True)
renameCols = [(col, 'swin_'+col) for col in swin_df.columns if col not in ['image_id']]
swin_df.rename(columns = dict(renameCols), inplace=True)

swin_df.head(3)
```

```python
eff3_df.rename(columns = {'Unnamed: 0':'image_id'}, inplace=True)
renameCols = [(col, 'eff3_'+col) for col in eff3_df.columns if col not in ['image_id']]
eff3_df.rename(columns = dict(renameCols), inplace=True)

eff3_df.head(3)
```

```python
effv2_df.rename(columns={'negative':'neg',
                         'typical':'typ',
                         'indeterminate':'ind',
                         'atypical':'atyp'}, inplace=True)

renameCols = [(col, 'effv2_'+col) for col in effv2_df.columns if col not in ['image_id']]
effv2_df.rename(columns = dict(renameCols), inplace=True)

effv2_df.head(3)
```

```python
effv2_tta1_df.rename(columns={'negative':'neg',
                         'typical':'typ',
                         'indeterminate':'ind',
                         'atypical':'atyp'}, inplace=True)
renameCols = [(col, 'effv2_tta1_'+col) for col in effv2_tta1_df.columns if col not in ['image_id']]
effv2_tta1_df.rename(columns = dict(renameCols), inplace=True)

display(effv2_tta1_df.head(2))


effv2_tta2_df.rename(columns={'negative':'neg',
                         'typical':'typ',
                         'indeterminate':'ind',
                         'atypical':'atyp'}, inplace=True)
renameCols = [(col, 'effv2_tta2_'+col) for col in effv2_tta2_df.columns if col not in ['image_id']]
effv2_tta2_df.rename(columns = dict(renameCols), inplace=True)

display(effv2_tta2_df.head(2))


effv2_tta4_df.rename(columns={'negative':'neg',
                         'typical':'typ',
                         'indeterminate':'ind',
                         'atypical':'atyp'}, inplace=True)
renameCols = [(col, 'effv2_tta4_'+col) for col in effv2_tta4_df.columns if col not in ['image_id']]
effv2_tta4_df.rename(columns = dict(renameCols), inplace=True)

display(effv2_tta4_df.head(2))


effv2_tta6_df.rename(columns={'negative':'neg',
                         'typical':'typ',
                         'indeterminate':'ind',
                         'atypical':'atyp'}, inplace=True)
renameCols = [(col, 'effv2_tta6_'+col) for col in effv2_tta6_df.columns if col not in ['image_id']]
effv2_tta6_df.rename(columns = dict(renameCols), inplace=True)

display(effv2_tta6_df.head(2))
```

# Combine OOF into single dataframe

```python
oof_df = pd.merge(efnb7_df, swin_df, on="image_id", how="left")
oof_df = pd.merge(oof_df, eff3_df, on="image_id", how="left")
oof_df = pd.merge(oof_df, effv2_df, on="image_id", how="left")
oof_df = pd.merge(oof_df, effv2_tta1_df, on="image_id", how="left")
oof_df = pd.merge(oof_df, effv2_tta2_df, on="image_id", how="left")
oof_df = pd.merge(oof_df, effv2_tta4_df, on="image_id", how="left")
oof_df = pd.merge(oof_df, effv2_tta6_df, on="image_id", how="left")

print(oof_df.shape)
display(oof_df.head(3))
```

# Calculate mAP

```python
oof_df.columns
```

```python
y_true = oof_df[['Negative for Pneumonia','Typical Appearance','Indeterminate Appearance','Atypical Appearance']].values
y_efnb7 = oof_df[['efnb7_neg','efnb7_typ','efnb7_ind','efnb7_atyp']].values
y_swin = oof_df[['swin_neg','swin_typ','swin_ind','swin_atyp']].values
y_eff3 = oof_df[['eff3_neg','eff3_typ','eff3_ind','eff3_atyp']].values
y_effv2 = oof_df[['effv2_neg','effv2_typ','effv2_ind','effv2_atyp']].values
y_effv2_tta1 = oof_df[['effv2_tta1_neg','effv2_tta1_typ','effv2_tta1_ind','effv2_tta1_atyp']].values
y_effv2_tta2 = oof_df[['effv2_tta2_neg','effv2_tta2_typ','effv2_tta2_ind','effv2_tta2_atyp']].values
y_effv2_tta4 = oof_df[['effv2_tta4_neg','effv2_tta4_typ','effv2_tta4_ind','effv2_tta4_atyp']].values
y_effv2_tta6 = oof_df[['effv2_tta6_neg','effv2_tta6_typ','effv2_tta6_ind','effv2_tta6_atyp']].values
```

```python
from sklearn.metrics import average_precision_score

mAPloss = average_precision_score(y_true, y_efnb7)
print(f"efnb7 mAP (2/3): {(2*mAPloss)/3}")

mAPloss = average_precision_score(y_true, y_swin)
print(f"swin mAP (2/3): {(2*mAPloss)/3}")

mAPloss = average_precision_score(y_true, y_eff3)
print(f"eff3 mAP (2/3): {(2*mAPloss)/3}")

mAPloss = average_precision_score(y_true, y_effv2)
print(f"effv2 mAP (2/3): {(2*mAPloss)/3}")

mAPloss = average_precision_score(y_true, y_effv2_tta1)
print(f"effv2 tta1 mAP (2/3): {(2*mAPloss)/3}")

mAPloss = average_precision_score(y_true, y_effv2_tta2)
print(f"effv2 tta2 mAP (2/3): {(2*mAPloss)/3}")

mAPloss = average_precision_score(y_true, y_effv2_tta4)
print(f"effv2 tta4 mAP (2/3): {(2*mAPloss)/3}")

mAPloss = average_precision_score(y_true, y_effv2_tta6)
print(f"effv2 tta6 mAP (2/3): {(2*mAPloss)/3}")
```

# Averaging | Weighted Averaging 

```python
print("Average")

y_avg = np.mean(np.array([y_efnb7, y_swin, y_eff3, y_effv2]), axis=0)
mAPloss = average_precision_score(y_true, y_avg)

print(f"Average mAP (2/3): {(2*mAPloss)/3}")

y_avg = np.mean(np.array([y_efnb7, y_swin, y_eff3, y_effv2_tta1]), axis=0)
mAPloss = average_precision_score(y_true, y_avg)

print(f"Average tta1 mAP (2/3): {(2*mAPloss)/3}")

y_avg = np.mean(np.array([y_efnb7, y_swin, y_eff3, y_effv2_tta2]), axis=0)
mAPloss = average_precision_score(y_true, y_avg)

print(f"Average tta2 mAP (2/3): {(2*mAPloss)/3}")

y_avg = np.mean(np.array([y_efnb7, y_swin, y_eff3, y_effv2_tta4]), axis=0)
mAPloss = average_precision_score(y_true, y_avg)

print(f"Average tta4 mAP (2/3): {(2*mAPloss)/3}")

y_avg = np.mean(np.array([y_efnb7, y_swin, y_eff3, y_effv2_tta6]), axis=0)
mAPloss = average_precision_score(y_true, y_avg)

print(f"Average tta6 mAP (2/3): {(2*mAPloss)/3}")
```

```python
print("Weighted Average")

y_avg = (y_efnb7 + (2*y_swin) + y_eff3 + (2*y_effv2)) / 6
mAPloss = average_precision_score(y_true, y_avg)

print(f"Weighted Average mAP (2/3): {(2*mAPloss)/3}")

y_avg = (y_efnb7 + (2*y_swin) + y_eff3 + (2*y_effv2_tta1)) / 6 
mAPloss = average_precision_score(y_true, y_avg)

print(f"Weighted Average tta1 mAP (2/3): {(2*mAPloss)/3}")

y_avg = (y_efnb7 + (2*y_swin) + y_eff3 + (2*y_effv2_tta2)) / 6
mAPloss = average_precision_score(y_true, y_avg)

print(f"Weighted Average tta2 mAP (2/3): {(2*mAPloss)/3}")

y_avg = (y_efnb7 + (2*y_swin) + y_eff3 + (2*y_effv2_tta4)) / 6
mAPloss = average_precision_score(y_true, y_avg)

print(f"Weighted Average tta4 mAP (2/3): {(2*mAPloss)/3}")

y_avg = (y_efnb7 + (2*y_swin) + y_eff3 + (2*y_effv2_tta6)) / 6
mAPloss = average_precision_score(y_true, y_avg)

print(f"Weighted Average tta6 mAP (2/3): {(2*mAPloss)/3}")
```

# Rank averaging | Weighted rank averaging

```python
y_rank_efnb7 = oof_df[['efnb7_neg','efnb7_typ','efnb7_ind','efnb7_atyp']].rank().values
y_rank_swin = oof_df[['swin_neg','swin_typ','swin_ind','swin_atyp']].rank().values
y_rank_eff3 = oof_df[['eff3_neg','eff3_typ','eff3_ind','eff3_atyp']].rank().values
y_rank_effv2 = oof_df[['effv2_neg','effv2_typ','effv2_ind','effv2_atyp']].rank().values
y_rank_effv2_tta1 = oof_df[['effv2_tta1_neg','effv2_tta1_typ','effv2_tta1_ind','effv2_tta1_atyp']].rank().values
y_rank_effv2_tta2 = oof_df[['effv2_tta2_neg','effv2_tta2_typ','effv2_tta2_ind','effv2_tta2_atyp']].rank().values
y_rank_effv2_tta4 = oof_df[['effv2_tta4_neg','effv2_tta4_typ','effv2_tta4_ind','effv2_tta4_atyp']].rank().values
y_rank_effv2_tta6 = oof_df[['effv2_tta6_neg','effv2_tta6_typ','effv2_tta6_ind','effv2_tta6_atyp']].rank().values
```
